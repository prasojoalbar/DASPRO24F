```python
 import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

```

```python
df = pd.read_csv("NBA_PLAYERS.csv")
print(df)
```

```python
mean_value = np.mean(df["Height"])
plt.figure(figsize=(8, 5))
plt.hist(df["Height"], bins=15 ,color='red', edgecolor='black' , alpha=0.5, label='Tinggi Badan')
plt.axvline(mean_value , color='green',linestyle='--', linewidth='2' , label=f'Mean: {mean_value:.2f}')
plt.title('rata rata tinggi pemain')
plt.xlabel('Tinggi')
plt.ylabel('Frekuensi')
plt.legend()
plt.show()
```

```python
x = df[["Height"]]
y = df["PTS"]

model = LinearRegression()
model.fit(x, y)

y_pred = model.predict(x)

plt.figure(figsize=(8, 6))
plt.scatter(df['Height'], y_pred, color='red', label='Titik Nilai')
plt.plot(df['Height'],y_pred, color='blue', label='Garis Regresi')
plt.title('Prediksi antara Tinggi badan Dan Point')
plt.xlabel('Tinggi Badan')
plt.ylabel('Banyak Point')
plt.legend()
plt.grid(True)
plt.show()
```
