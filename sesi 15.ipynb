```python
#MAtplotlib, seaborn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(42)

bulan = ['januari','februari','maret','april','mei','juni','juli','agustus','september','oktober','november','desember']
Minggu_1 = np.random.randint(200000,700001,12)
Minggu_2 = np.random.randint(200000,700001,12)
Minggu_3 = np.random.randint(200000,700001,12)
Minggu_4 = np.random.randint(200000,700001,12)

data = {
    'bulan' : bulan ,
    'Minggu1' : Minggu_1,
    'Minggu2' : Minggu_2,
    'Minggu3' : Minggu_3,
    'Minggu4' : Minggu_4
}
df = pd.DataFrame(data)
df['Total Pengeluaran']= df[['Minggu1','Minggu2','Minggu3','Minggu4']].sum(axis=1)
df
```

```python
##diagram bar
#buat canvas
plt.figure(figsize=(12,6)) #widht,height -> inch
plt.bar(df['bulan'],df['Total Pengeluaran'],color='magenta')
plt.xticks(rotation=45)
plt.tight_layout()
plt.title("Total Pengeluaran Mahasiswa Per Bulan")
plt.ylabel("Rupiah")
plt.show()
```

```python
#tren pengeluaran per minggu dalam satu tahun

plt.figure(figsize=(10,10))

for minggu in ['Minggu1','Minggu2','Minggu3','Minggu4']:
    plt.plot(df['bulan'],df[minggu],marker='o',label=minggu)

plt.title("Tren Pengeluaran Mahasiswa Per Minggu")
plt.ylabel("Rupiah")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#diagram pie (proporsi pengeluaran perbulan)
plt.figure(figsize=(6,6))
plt.pie(df['Total Pengeluaran'],labels=df['bulan'], startangle=90 , autopct='%1.1f%%', colors=plt.cm.Pastel2.colors)
plt.axis('equal')
plt.title("Proporsi Pengeluaran Mahasiswa Per Bulan")
plt.show()
```

```python
#Model AI  -> scikit , sklearn
from sklearn.linear_model import LinearRegression
df['bulan'] = np.arange(1,13)

X = df[['bulan']]
Y = df[['Total Pengeluaran']]

model = LinearRegression()
model.fit(X,Y)

prediksi_13 = model.predict([[13]])[0]
print(f"Prediksi Pengeluaran Mahasiswa Bulan Januari:{float(prediksi_13):,.0f}")
plt.figure(figsize=(10,10))
plt.boxplot(df['Total Pengeluaran'])
plt.show()
```
