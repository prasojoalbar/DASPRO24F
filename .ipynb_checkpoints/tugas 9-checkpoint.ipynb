```python
import numpy as np
import pandas as pd

n_data = 10000


usia = np.random.randint(22, 40, size=n_data)
jenis_kelamin = np.random.choice(['Laki-laki', 'Perempuan'], size=n_data)
pendidikan = np.random.choice(['SD', 'SMP', 'SMA', 'Diploma', 'Sarjana', 'Pascasarjana'], size=n_data)
pekerjaan = np.random.choice([ 'Guru', 'Mahasiswa', 'Kuli', 'PNS', 'Pelajar', 'Pengangguran'], size=n_data)
status = np.random.choice(['Belum Kawin', 'Kawin', 'Cerai'], size=n_data)
jumlah_anak = np.random.randint(0, 6, size=n_data)
Gaji = np.random.randint(500_000, 20_000_000, size=n_data)
kota = np.random.choice([ 'Jampang', 'Cisaat', 'Cibadak', 'Pelabuhan Ratu', 'Parungkuda', 'Sukabumi'], size=n_data)
nik = [''.join(np.random.choice(list('0123456789'), size=16)) for _ in range(n_data)]

df = pd.DataFrame({
    
    'Usia': usia,
    'Jenis_Kelamin': jenis_kelamin,
    'Pendidikan_Terakhir': pendidikan,
    'Pekerjaan': pekerjaan,
    'Status_Perkawinan': status,
    'Jumlah_Anak': jumlah_anak,
    'Penghasilan_Bulanan': Gaji,
    'Kota_Tinggal': kota,
    'NIK': nik
})


df.to_csv('data_sensus.csv', index=False)


df.head()
```
